# Appendix 11: Hypothesis 3 Test

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import statsmodels.api as sm
from scipy import stats

from statsmodels.formula.api import ols
```

```python
df=pd.read_csv('data_final.csv')
df.head()
```

```python
df=df.iloc[:,1:]
df.head()
```

```python
genres=df['genres'].unique()
decades=df['Decade'].unique()
print(genres)
print(decades)
```

```python
tb = [[0 for x in range(6)] for y in range(len(df))]
len(tb)
```

```python
df_1=df
y=df_1.iloc[:,[25,26]]
x=df_1.iloc[:,[3,4,10,12]]
norm = MinMaxScaler().fit(x)
x_norm = norm.transform(x)
x=pd.DataFrame(data=x_norm,columns=x.columns)
x.head()
data=pd.concat([x,y],axis=1)
data.head()
```

```python
for index,row in data.iterrows():
    tb[index][0]=row['genres']
    tb[index][1]=row['Decade']
    tb[index][2]=row['danceability']
    tb[index][3]=row['energy']
    tb[index][4]=row['instrumentalness']
    tb[index][5]=row['valence']
```

```python
tb[:5]
```

```python
x=['Genre','Decade','Danceability','Energy','Instrumentalness','Valence']
df_tb=pd.DataFrame(data=tb,columns=x)
df_tb.head()
```

```python
d_melt=df_tb.iloc[:,:3]
d_melt.head()
```

```python
model = ols('Danceability ~ C(Genre) + C(Decade) + C(Genre):C(Decade)', data=d_melt).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
anova_table
```

```python
d_melt=df_tb.iloc[:,[0,1,3]]
model = ols('Energy ~ C(Genre) + C(Decade) + C(Genre):C(Decade)', data=d_melt).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
anova_table
```

```python
d_melt=df_tb.iloc[:,[0,1,4]]
model = ols('Instrumentalness ~ C(Genre) + C(Decade) + C(Genre):C(Decade)', data=d_melt).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
anova_table
```

```python
d_melt=df_tb.iloc[:,[0,1,5]]
model = ols('Valence ~ C(Genre) + C(Decade) + C(Genre):C(Decade)', data=d_melt).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
anova_table
```
